```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data_df["SN"].value_counts())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Pulls unique item names and counts them
unique_items= purchase_data_df["Item Name"].unique()
total_items = len(unique_items)

#This code simply averages all the prices in the list
average = purchase_data_df["Price"].mean()

#Counts the total number of purchases
total_purchases = len(purchase_data_df["Purchase ID"])

#Calculates the sum of the price column for total revenue
total_revenue = purchase_data_df["Price"].sum()

#Creates the summary dataframe
purchasing_summary_df = pd.DataFrame({"Number of Unique Items": [total_items],
                                      "Average Price": [average], "Number of Purchases": [total_purchases],
                                      "Total Revenue": [total_revenue]})
#Displays the new dataframe
purchasing_summary_df

#How do I format the price columns for dollar signs??
purchasing_summary_df_formatted = purchasing_summary_df.style.format({"Average Price":"${:,.3}".format,
                                                                     "Total Revenue":"${:,}".format})
purchasing_summary_df_formatted
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Only pull male players
males_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]

#Count the number of males using the unique SN
total_males = len(males_df["SN"].unique())
total_males

#Only pull female players
females_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
total_females = len(females_df["SN"].unique())
total_females

#Calculate other/non
other = total_players - total_males - total_females
other

#Calculate percent of males
percent_male = total_males/total_players 

#Calculate percent of females
percent_female = total_females/total_players

#Calculate percent of other
percent_other = other/total_players

#Create a summary dataframe
gender_summary_df = pd.DataFrame({"Total Count": [total_males, total_females, other], "Percentage of Players":
                                [percent_male, percent_female, percent_other]})

#Change the index labels to match data
gender_summary_rename = gender_summary_df.rename(index={0:"Male", 1:"Female", 2:"Other/Non-Discolosed"})

#Change the formatting for percentage and two decimal places
gender_summary_df_formatted = gender_summary_rename.style.format({"Percentage of Players":"{:,.2%}".format })

gender_summary_df_formatted
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Accurately counts number of purchases per gender
gender_purchase_totals = purchase_data_df["Gender"].value_counts()
#Changes it to a dataframe and adds index title for first column
gender_purchase_totals_df = pd.DataFrame(gender_purchase_totals).reset_index()
gender_purchase_totals_df

#Define variable for total male purchases for summary df
male_purchases = gender_purchase_totals_df.iloc[0,1]

#Define variable for total female purchases for summary df
female_purchases = gender_purchase_totals_df.iloc[1,1]

#Define variable for total other purchases for summary df
other_purchases = gender_purchase_totals_df.iloc[2,1]



#Lists the data by gender and price only
purchase_price_gender_df = purchase_data_df[["Gender", "Price"]]

#Calculates a table with gender and price for each already averaged
average_price_gender_df = purchase_price_gender_df.groupby("Gender").mean().reset_index()
average_price_gender_df

#Define variable for avg purchase price by gender
female_avg_price = average_price_gender_df.iloc[0,1]
male_avg_price = average_price_gender_df.iloc[1,1]
other_avg_price = average_price_gender_df.iloc[2,1]


#Calculates a table with gender and sum of purchases for each gender
sum_purchases_gender_df = purchase_price_gender_df.groupby("Gender").sum().reset_index()
sum_purchases_gender_df

#Define variable for sum of total purchases by gender
female_total_purchases = round(sum_purchases_gender_df.iloc[0,1],2)
male_total_purchases = round(sum_purchases_gender_df.iloc[1,1],2)
other_total_purchases = round(sum_purchases_gender_df.iloc[2,1],2)

#Calculate total purchase per person by gender
avg_per_male = sum_purchases_gender_df.iloc[1,1]/total_males
avg_per_female = sum_purchases_gender_df.iloc[0,1]/total_females
avg_per_other = sum_purchases_gender_df.iloc[2,1]/other

#Create summary df
summary_purchasing_gender_df = pd.DataFrame({"Gender": ["Female", "Male", "Other/Non-Disclosed"],"Purchase Count":
                                             [female_purchases, male_purchases, other_purchases], "Average Purchase Price":
                                [female_avg_price, male_avg_price, other_avg_price], "Total Purchase Value":
                                             [female_total_purchases, male_total_purchases, other_total_purchases],
                                             "Avg Total Purchase per Person": [avg_per_female, avg_per_male, avg_per_other]})

#Set Gender as the index
index_summary_purchasing_gender_df = summary_purchasing_gender_df.set_index("Gender")
index_summary_purchasing_gender_df

#Format the columns with price data with $ and two decimal places
formatted_summary_purchasing_gender_df = index_summary_purchasing_gender_df.style.format({"Average Purchase Price":"${:,.3}".format,
                                                                     "Total Purchase Value":"${:,}".format,
                                                                                          "Avg Total Purchase per Person":"${:,.3}"})
formatted_summary_purchasing_gender_df

#Just need to format the female average purchase price - how do i get that zero??
```

```python
#Group the purchase data by SN and price, with SN being first to align each purchase with the unique SN
SN_purchase_data = purchase_data_df.groupby(["SN", "Price"])

#Use a count function to be able to visualize the results
SN_purchase_data_count = SN_purchase_data.count()

#Make the grouped data into a dataframe
SN_purchase_data_df = pd.DataFrame(SN_purchase_data_count)
SN_purchase_data_df

#Delete Unnecessary Columns from the grouped SN/price data dataframe
del SN_purchase_data_df["Purchase ID"]
del SN_purchase_data_df["Age"]
del SN_purchase_data_df["Item ID"]

#Reset the index of the data in the grouped SN/price dataframe so that the column names can be referenced 
reset_df = SN_purchase_data_df.reset_index()

#Only group the new dataframe by the unique SN and sum the remaining columns - this will calculate total purchase value AND number of purchases made per person
grouped_data = reset_df.groupby(["SN"]).sum()
grouped_data

#Make a dataframe from the grouped SN/sum information and reset the index so the column names can be referenced
grouped_data_df = pd.DataFrame(grouped_data)
grouped_data_df_reset = grouped_data_df.reset_index()
grouped_data_df_reset

#Sorts the data in the grouped SN/sum information dataframe by price, descending to show the top spenders
sorted_grouped_data_df = grouped_data_df_reset.sort_values("Price", ascending = False)
sorted_grouped_data_df

#Reorganizes the columns fot the summary dataframe
organized_SN_purchase_df = sorted_grouped_data_df[["SN", "Item Name", "Price"]]
renamed_SN_purchase_df = organized_SN_purchase_df.rename(columns={"Item Name": "Purchase Count", "Price": "Total Purchase Value"})
renamed_SN_purchase_df

#Assign a variable and calculates the average purchase price per person
avg_purchase_price = renamed_SN_purchase_df["Total Purchase Value"]/renamed_SN_purchase_df["Purchase Count"]

#Add a column to the summary dataframe to reflect the avg purchase price per person
renamed_SN_purchase_df["Average Purchase Price"] = avg_purchase_price

#Switch the positions of total value and average price
final_org_SN_purchase_df = renamed_SN_purchase_df[["SN", "Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
final_org_SN_purchase_df

#Set SN as the index
SN_index_purchase_df = final_org_SN_purchase_df.set_index("SN").head()

#Format the columns with price to show $ and two decimal places
SN_final_purchase_df = SN_index_purchase_df.style.format({"Average Purchase Price":"${:,.3}".format,
                                                                     "Total Purchase Value":"${:,.4}".format})
SN_final_purchase_df

#Again, how do i get that zero on the 5th line down?? Also, why after formatting does it show the entire data set?
#It's not acting like a df, listing rows and columns at bottom and allowing just a header.
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get Item ID, Item Name and Price
most_popular_df = purchase_data_df[["Item ID", "Item Name", "Price"]]

#Group by Item ID first then Item Name
popular_grouped = most_popular_df.groupby(["Item ID", "Item Name"])

#sum function to view the grouped data and obtain total purchase value for each item
item_price = popular_grouped.sum()

#convert to a dataframe so that the price values can be stored
item_price_df = pd.DataFrame(item_price)

#Reset the index so that the price column can be accessed
item_price_index = item_price_df.reset_index()

#Start creating the summary dataframe, based on the item price dataframe above
popular_summary_df = item_price_index.rename(columns={"Price": "Total Purchase Value"})

#Use the count function to obtain how many were purchased
purchase_count = popular_grouped.count()
#Convert to a dataframe so the number of purchases can be stored
purchase_count_df = pd.DataFrame(purchase_count)
#Reset the index so the columns can be accessed
purchase_count_index_df = purchase_count_df.reset_index()
#Set a variable to hold purchase count for the summary table
purchase_count_final = purchase_count_index_df["Price"]

#Get the average item price using the mean function
average_item_price = popular_grouped.mean()
#Convert to a dataframe so the average item price can be stored
average_item_price_df = pd.DataFrame(average_item_price)
#Reset the index so the columns can be accessed
average_item_price_index_df = average_item_price_df.reset_index()
#Set a variable to hold purchase count for the summary table
average_price_final = average_item_price_index_df["Price"]


#add the purchase count and average item price columns to the summary dataframe
popular_summary_df["Purchase Count"] = purchase_count_final
popular_summary_df["Item Price"] = average_price_final

#Rearrange columns in summary dataframe
popular_summary_df_organized = popular_summary_df[["Item ID", "Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]

#Sort purchase count column in descending order
popular_summary_df_sorted = popular_summary_df_organized.sort_values("Purchase Count", ascending = False)
popular_summary_df_sorted

#Set Item ID as Index
popular_summary_df_item_index = popular_summary_df_sorted.set_index("Item ID")
popular_summary_df_item_index

#Format the price columns to have $ and 2 decimals
popular_summary_df_final = popular_summary_df_item_index.style.format({"Item Price":"${:,.3}".format,
                                                                     "Total Purchase Value":"${:,.4}".format})

popular_summary_df_final
#Again can't display first five after doing the formatting on the dataframe. Why??
```
